```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/blood-transfusion/transfusion.data"
df  = pd.read_csv(url)
```

```python
df.shape
```

```python
df.info()
```

```python
df.columns
```

```python
df.columns = ['Recency (months)', 'Frequency (times)', 'Monetary (c.c. blood)',
       'Time (months)', 'Donated Blood(Y/N)']
```

```python
df['Donated Blood(Y/N)'].value_counts()
```

```python
fig, axes = plt.subplots(1, 4, figsize=(27, 5))
fig.suptitle('Blood Transfusion Box Plot Result')

sns.boxplot(ax=axes[0], x='Donated Blood(Y/N)', y='Recency (months)', data=df)

sns.boxplot(ax=axes[1], x='Donated Blood(Y/N)', y='Frequency (times)', data=df)

sns.boxplot(ax=axes[2], x='Donated Blood(Y/N)', y='Monetary (c.c. blood)', data=df)

sns.boxplot(ax=axes[3], x='Donated Blood(Y/N)', y='Time (months)', data=df)
plt.show()
```

```python
sns.pairplot(df[['Recency (months)', 'Frequency (times)', 'Monetary (c.c. blood)',
       'Time (months)', 'Donated Blood(Y/N)']],hue='Donated Blood(Y/N)')
plt.show()
```
